This is a machine learning model based on multiple linear regression .
In the dataset we try to predict which product can be pitched to the user based on the prediction made by the model based on the previous data.
The dataset consists of various values like 'Age','Type of contact','City Tier' and many more.
First we import the libraries needed to run the model like pandas,pylab and numpy.

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

Using the pd.read_csv command we import the file on which we will make prediction, after that we use head command to print the first few rows of our model.

```python
df=pd.read_csv(r'C:/Users/yashi/OneDrive/Desktop/mycodes/hotel_data.csv')
df.head(12)
```

```python
cdf=df[['Age','TypeofContact','CityTier','Occupation','MaritalStatus','NumberOfTrips','Passport','ProductPitched']]
cdf.head(10)
cdf=df[['TypeofContact','CityTier','Occupation','MaritalStatus','NumberOfTrips','Passport','ProductPitched']].values
cdf[0:9]
from sklearn import preprocessing

le_TypeofContact=preprocessing.LabelEncoder()
le_TypeofContact.fit(['Company Invited','Self Enquiry'])
cdf[:,0]=le_TypeofContact.transform(cdf[:,0])

le_Occupation=preprocessing.LabelEncoder()
le_Occupation.fit(['Salaried','Free Lancer','Small Business','Large Business'])
cdf[:,2]=le_Occupation.transform(cdf[:,2])



le_MaritalStatus=preprocessing.LabelEncoder()
le_MaritalStatus.fit(['Single','Married','Unmarried','Divorced'])
cdf[:,3]=le_MaritalStatus.transform(cdf[:,3])

le_ProductPitched=preprocessing.LabelEncoder()
le_ProductPitched.fit(['Basic','Standard','King','Deluxe','Super Deluxe'])
cdf[:,6]=le_ProductPitched.transform(cdf[:,6])
```

```python
msk=np.random.rand(len(cdf))<0.8
train=cdf[msk]
test=cdf[~msk]
```

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
x=np.asanyarray(train[:,[0,1,2,3,4,5]])
y=np.asanyarray(train[:,[6]])
regr.fit(x,y)
print('coefficients',regr.coef_)
```

```python
y_hat=regr.predict(test[:,[0,1,2,3,4,5]])
x=np.asanyarray(test[:,[0,1,2,3,4,5]])
y=np.asanyarray(test[:,[6]])
```

```python
def predict_services(data):
    assert isinstance(data, list) and len(data) == 6
    
   
    data = np.array(data).reshape(1, -1)
    
    
    predicted_service = regr.predict(data)
    
   
    predicted_service = le_ProductPitched.inverse_transform(predicted_service.astype(int))
    
    return predicted_service[0]


input_data = [1, 1, 1, 1, 0, 0] 
print("Predicted service:", predict_services(input_data))

```
